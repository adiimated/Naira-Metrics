```python
import pandas as pd
import numpy as np
```

Reading Money Supply

```python
money_supply = pd.read_excel('Money Supply 1960-2024.xlsx')
```

```python
money_supply.shape
```

```python
money_supply.head(3)
```

```python
money_supply.isna().sum()
```

```python
money_supply['Money Supply M3'].fillna(value=money_supply['Money Supply M3'].mean(), inplace=True)

money_supply['Credit To GovernmentFed'].fillna(value=money_supply['Credit To GovernmentFed'].mean() , inplace=True)

money_supply['Mirror Accounts'].fillna(value=money_supply['Mirror Accounts'].mean(), inplace=True)

money_supply['CBN Bills'].fillna(value=money_supply['CBN Bills'].mean(), inplace=True)

money_supply['Special Intervention Reserves'].fillna(value=money_supply['Special Intervention Reserves'].mean(), inplace=True)
```

```python
money_supply.isna().sum()
```

A discrepency has been identified in the Months column of given data. For rowID 375 onward, the Month value is set to year value. The is needed to be correct for future analysis.

```python
money_supply.tail(35)
```

The identified problem in given data is resolved by using Group by function which is used to group data based on year and computes mean for months value ( if available)

```python
money_supply = money_supply.groupby(['Year'], as_index=True).mean().round(2)
```

The month column has beed dropped as we longer need it

```python
money_supply.drop(columns='Month', inplace=True)
money_supply.head(5)
```

```python
money_supply.duplicated().unique()
```

```python
money_supply.to_csv('MoneySupply_1960-2024_Yearly.csv')
```
