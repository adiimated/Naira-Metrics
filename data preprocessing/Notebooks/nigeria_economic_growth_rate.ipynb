```python
import pandas as pd
import numpy as np
```

```python
url = "https://docs.google.com/spreadsheets/d/1y-Uk3I-NroVSNVETvwCn2HPq0YiAmBafoO9J5SqsOg4/export?format=csv&gid=899151651"
pipe_original = pd.read_csv(url)
data = pipe_original.copy()
data.head()
```

```python
df = data.copy()
df.iloc[10:20]
```

## Drop unnecessary rows

```python
df = df.iloc[14:]
df
```

# Trying to merge first 2 rows

```python
new_row = df.iloc[0] + df.iloc[1]
```

```python
new_row
```

```python
df = pd.concat([pd.DataFrame([new_row]), df]).reset_index(drop=True)
df
```

```python
df = df.drop([1,2])
df.columns = df.iloc[0]
df = df.iloc[1:].reset_index(drop=True)
df
```

```python
df.columns
```

```python
df.columns = ['Year', 'GDPBillions of US $', 'Per CapitaUS $', 'Growth RateAnnual % Change']
```

#Extracting Year from first Row

```python
df['Year'] = df['Year'].astype(str).str[:4]
```

```python
df.head()
```

```python
df = df.set_index('Year')
df.head()
```
