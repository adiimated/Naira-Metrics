```python
# load dataset
# drop unnecessary columns
# Reshape/melt the DataFrame , where each row represents a combination of 'Name of Series', 'Year', and 'Value'.
#using pivot() function on df_transformed to pivot the data, setting 'Year' as the index,'Name of Series' as columns, and 'Value' as values.
#This reshapes the DataFrame so that each series ('Crop production', 'Livestock', 'Forestry', 'Fishing') has its own column with values across years.
```

```python
import pandas as pd
import numpy as np
#load data 
df=pd.read_csv('Agriculture Anually( Crop Production, Livestock, Forestery, Fishing).csv')
df
```

```python
#drop uncessery columns
cols_drop=['Mnemonic','Parent Mnemonic','Databank','Data Family','Data Source','Units of Measure','First Observation','Frequency']
df=df.drop(cols_drop,axis =1)
df
```

```python
# Reshape/melt the DataFrame , where each row represents a combination of 'Name of Series', 'Year', and 'Value'.

df_transformed = df.melt(id_vars='Name of Series', var_name='Year', value_name='Value')
df_transformed 
```

```python
# Pivot the DataFrame setting 'Year' as the index, 'Name of Series' as columns, and 'Value' as values.
#This reshapes the DataFrame so that each series ('Crop production', 'Livestock', 'Forestry', 'Fishing') 
#has its own column with values across years.
df=df_transformed.pivot(index='Year',columns='Name of Series',values='Value')
df.columns=['Crop production', 'Livestock', 'Forestry', 'Fishing']
df.reset_index()
```

```python
# save dataset to csv file
df.to_csv('Agriculture_Anually.csv')
```
