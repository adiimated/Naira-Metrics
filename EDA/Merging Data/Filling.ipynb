```python
import pandas as pd

# Load the time series data from the uploaded file
file_path = '1960 onwards.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data to understand its structure
data.head()
```

```python
# Applying interpolation first
data_interpolated = data.interpolate(method='linear')

# Applying forward fill and backward fill
data_filled = data_interpolated.ffill().bfill()

# Save the filled dataset to a new CSV file
output_file_path = '1960_onwards_filled.csv'
data_filled.to_csv(output_file_path, index=False)
```

```python
import pandas as pd

# Load the new time series data
new_file_path = '1980 Onwards.csv'
new_data = pd.read_csv(new_file_path)

# Applying backward fill to handle missing values
new_data_filled = new_data.bfill()

# Save the filled dataset to a new CSV file
output_file_path_new = '1980_Onwards_filled.csv'
new_data_filled.to_csv(output_file_path_new, index=False)
```
